```r
#
#     E3asknsn a, b, g
#
#####################################################

n<-100
U<-runif(n)
cat("E(U)=",0.5,"var(U)=",1/12,"Median=",0.5,"\n")
c(mean(U),var(U),median(U))
hist(U,nclass=25,col=4,xlab="x",main="Histogram of U(0,1)",
     xlim=c(-0.1,1.1))
hist(U,col=4)


X<--2*U+1
cat("E(X)=",-2*0.5+1,"var(X)=",(-2)^2/12,"Median=",-2*0.5+1,"\n")
c(mean(X),var(X),median(X))
hist(X,nclass=25,col=4,xlab="x")
```

```r
#
#     E3asknsn d, e
#
#####################################################

n<-1000
U<-runif(n)
cat("E(U)=",0.5,"var(U)=",1/12,"Median=",0.5,"\n")
c(mean(U),var(U),median(U))
hist(U,nclass=25,col=4,main="Histogram of U(0,1)")
for (th in c(0.8,3,10)){
Y<--th*log(1-U)
cat("E(Y)=",th,"var(Y)=",th^2,"Median=",-th*log(0.5),"\n")
cat(mean(Y),var(Y),median(Y),"\n")
hist(Y,nclass=25,col=4,main=paste("Histogram of Exp(",th,")"))
}
```

```r
n<-100
Z<-rnorm(n)
cat("E(Z)=",0,"var(Z)=",1,"Median=",0,"\n")
c(mean(Z),var(Z),median(Z))
hist(Z,nclass=25,col=4,main="Histogram of N(0,1)")
W<-pnorm(Z)
cat("E(W)=",0.5,"sd(W)=",sqrt(1/12),"\n")
cat(mean(W),sd(W),"\n")
hist(W,nclass=25,col=4,main=paste("Histogram of U(0,1)"))

quantile(Z,  probs = c(5, 95)/100)
qnorm(c(5, 95)/100)
c(sort(Z)[5],sort(Z)[95])
c(sort(W)[5],sort(W)[95])
sort(W)[5]==pnorm(sort(Z)[5])
sort(W)[95]==pnorm(sort(Z)[95])
```

```r
install.packages('ggExtra')

M<-1000
x1<-rnorm(M)
x2<-rnorm(M)
x3<-rnorm(M)
x4<-rnorm(M)
x<-data.frame(x1,x2,x3,x4)
# x
apply(x,1,mean)
apply(x,2,mean)

Z<-apply(x,1,mean)
summary(Z)
var(Z)

hist(Z)
hist(Z,breaks=50,probability=TRUE)
curve(dnorm(x,mean=0,sd=1/sqrt(4)),add=TRUE,col=3,lwd=2)

S2<-apply(x,1,var)
summary(3*S2)
var(3*S2)


hist(3*S2,breaks=50,probability=TRUE)
curve(dchisq(x,df=3),add=TRUE,col=3,lwd=2)

T<-apply(x,1,mean)*sqrt(4)/apply(x,1,sd)
summary(T)
var(T)

hist(T,breaks=200,probability=TRUE)
curve(dt(x,df=3),add=TRUE,col=3,lwd=2)

M<-1000
U1<-runif(M)
U2<-runif(M)

R<-sqrt(-2*log(U1))
Theta<-2*pi*U2
Z1<-R*cos(Theta)
Z2<-R*sin(Theta)
Z<-data.frame(Z1,Z2)


cor(Z)

library(ggplot2)
library(ggExtra)

p<-ggplot(Z, aes(x=Z1, y=Z2))+
  geom_point() +
  theme_classic()

ggMarginal(p, type="histogram",bins=30,xparams=list(colour="blue",fill="red"))
```
