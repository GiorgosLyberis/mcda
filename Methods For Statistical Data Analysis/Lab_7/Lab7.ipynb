```r
x<-c(74, 63, 68, 71, 70, 71, 65, 71, 68, 73)


x
summary(x)
sort(x)
order(x)
rank(x)

sort(x)
rank(x[order(x)])

sort(x)
sort(x-65)
rank((x-65)[order(x)])
rank(abs(65-x)[order(x)])
rank(abs(x-65)[order(x)])
sign((x-65)[order(x)])

sign.test<-function(x,p,x0){
    y<-sum(x<x0)
    n<-length(x)
    return(binom.test(y,n,p))}

sign.test(x,0.25,65)

sign.test(x,0.5,69)
wilcox.test(x,mu=69)

sum(sign(x)*rank(abs(x)))
wilcox.test(x,alternative = "l")
wilcox.test(x,alternative = "g")
wilcox.test(x)

wilcox.test(x,mu=69)   # mu is better not to be an observation

sort(x)
sign(x-69)[order(x)]
rank(abs(x-69))[order(x)]
sign(x-69)*rank(abs(x-69))[order(x)]
sum(sign(x-69)*rank(abs(x-69)))
sum(rank(abs(x-69))[sign(x-69)>0])-sum(rank(abs(x-69))[sign(x-69)<0])
sum(rank(abs(x-69))[sign(x-69)>0])+sum(rank(abs(x-69))[sign(x-69)<0]) #n*(n+1)/2
sum(rank(abs(x-69))[sign(x-69)>0])


#
#p-value with ties  # in most of the cases
#
mu0<-69
wilcox.test(x,mu=mu0,correct=F)
T<-sum(rank(abs(x-mu0))*sign(x-mu0))/sqrt(sum(rank(abs(x-mu0))^2))
T
2*min(1-pnorm(T),pnorm(T))
#
#p-value without ties # by jittering data
#
nx<-length(x)
curve(psignrank(x,n=nx),xlim=c(-1,(nx*(nx+1)/2+1)))
hist(rsignrank(1000, n=nx),breaks=seq(-0.5,(nx*(nx+1)/2+.5),1))
x1<-jitter(x,factor=1/4)  # mikpn diatapaxn tou x gia ties
wilcox.test(x1,mu=mu0)
sum(rank(abs(x1-mu0))[sign(x1-mu0)>0])
V<-unname(wilcox.test(x1,mu=mu0)$stat)
V
2*min(1-psignrank(V-1,n=nx),psignrank(V,n=nx))
```
