```r
age<-rep(c("<30","30-50",">50"),c(7,9,4))
age
age<-c(age,rep(c("<30","30-50",">50"),c(1,10,0)))
age
age<-c(age,rep(c("<30","30-50",">50"),c(4,10,4)))
age<-c(age,rep(c("<30","30-50",">50"),c(5,27,1)))
age
age<-factor(age,levels=c("<30","30-50",">50"))
sex<-rep(c("M","F"),c(20,11))
sex<-c(sex,rep(c("M","F"),c(18,33)))
sex<-factor(sex,levels=c("M","F"))
sex
type<-rep(c("Serious","Minor"),c(31,51))
injury<-data.frame(age,sex,type)
injury
with(injury, table(age, sex, type))

#injury<-cbind(age,sex,type)
#dimnames(injury) <- list(1:82, c("age","sex","type"))

par(mfrow=c(2,2))
x<-table(injury)[,,1]
barplot(x,legend = TRUE,ylim = c(0, 80))
x<-t(table(injury)[,,1])
barplot(x,legend = rownames(x))
x<-table(injury)[,,2]
barplot(x,legend = rownames(x),ylim = c(0, 30),col=c("lightblue", "mistyrose",
                "lightcyan"), beside = TRUE)
x<-t(table(injury)[,,2])
barplot(x,legend = rownames(x),col=c("blue","pink"), beside = TRUE)
par(mfrow=c(1,1))
```

```r
f<-c(5,11,20,9,13,7)
x<-seq(5,55,10)
y<-rep(x,f)
hist(y, breaks = c(seq(0,60,10)),labels =as.character(f),freq = TRUE)
hist(y, breaks = c(seq(0,60,10)),labels =as.character(round(f/sum(f),2)),freq = FALSE)

p<-hist(y, breaks = c(seq(0,60,10)),labels =as.character(f),xaxt = "n",freq = TRUE,col = "lightblue")
p
axis(1, c(seq(0,60,10)))
lines(p$mids,p$counts,lwd=2)

hist(y, breaks = c(seq(0,60,10)),xaxt = "n",freq = FALSE,col = "lightblue")
axis(1, c(seq(0,60,10)))
lines(density(y),type="l",lwd=3)

hist(y, breaks = c(seq(0,60,10)),xaxt = "n",freq = TRUE,col = "lightblue")
axis(1, at = c(seq(0,60,10)), labels = FALSE,tick = TRUE)
axis(1, c(seq(5,55,10)), c("   0-9","   10-19","   20-29","   30-39",
	"   40-49","   50-59"), col.axis = "blue", tick = FALSE)
```

```r
x<-c(2,3,2,8,7,60,5,3,4)
sort(x)
mean(x)
mean(x, trim = 0.05)
mean(x, trim = 0.11)
mean(x, trim = 1/length(x))
mean(x, trim = 2/length(x))
mean(x, trim = 3/length(x))

wt <- c(rep(1,length(x)-1),  0)/(length(x)-1)
wt
sort(x)
weighted.mean(sort(x), wt)
wt <- c(0,rep(1,length(x)-2),  0)/(length(x)-2)
wt
weighted.mean(sort(x), wt)
```
