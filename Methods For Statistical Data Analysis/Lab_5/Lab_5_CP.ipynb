```r
#  Example 1
#
x<-c(0.54, 0.72, 0.85, 0.41, 0.22, 0.67, 0.78, 0.90, 0.64, 0.95)

ecdf(x)
plot(ecdf(x),do.points=FALSE, verticals=TRUE)
curve(pbeta(x,shape1=2, shape2=1), add=TRUE, xlim=c(0,1),col="red")

ks.test(x,"pbeta",shape1=2, shape2=1)


n<-length(x)
order.x<-sort(x)
ecdf(x)(order.x)
Fn<-ecdf(x)(order.x)
Fn
F0<-pbeta(order.x,shape1=2, shape2=1)
cbind(order.x,Fn,F0,Fn-F0,F0-c(0,Fn[-n]))
```

```r
#  Example 2
#

x<-c(18.8,19.3,22.4,22.5,24.0,24.7,25.9,27.0,35.1,35.8,36.5,37.6)

ecdf(x)
plot(ecdf(x),do.points=FALSE, verticals=TRUE)
curve(pnorm(x,mean=30,sd=sqrt(100)), add=TRUE, xlim=c(0,50),col="red")

ks.test(x,"pnorm",mean=30, sd=sqrt(100))

n<-length(x)
order.x<-sort(x)
ecdf(x)(order.x)
Fn<-ecdf(x)(order.x)
Fn
F0<-pnorm(order.x,mean=30, sd=sqrt(100))
cbind(order.x,Fn,F0,Fn-F0,F0-c(0,Fn[-n]))
```
