```r
#  posostiaia snmeia X^2 df=3, 10, 20

#df=3
v<-3
pchisq(-1,df=v)
pchisq(c(1,3,9),v)
qchisq(0.05,v)
qchisq(0.5,v)  #median
qchisq(0.95,v)

#df=10
v<-10
pchisq(c(1,10,23),v)
qchisq(0.05,v)
qchisq(0.5,v)  #median
qchisq(0.95,v)

#df=20
v<-20
qchisq(0.05,v)
pchisq(c(1,20,36),v)
qchisq(0.5,v)   #median
qchisq(0.95,v)
```
